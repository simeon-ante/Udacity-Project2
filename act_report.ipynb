As part of my Data Analysis Nano-degree program, Udacity has asked me to gather and analyse tweets from this twitter account. This was a really exciting project for me and I have shared my findings below

#### Dog Breeds

Be honest. How many breeds of Dogs do you know? I bet you don't know up to 112 breeds!
I found 112 different dog breeds in the WeRateDogs dataset.

Analysis showed that the most popular breeds (based on the number of twwets related to them) are the golden retreiver, labrador retreiver, pembroke, chihuahua and the pug in that order. 




From the chart produced, the most popular dog breed based on number of tweets about it is the **Golden Retriever with 139 tweets**

When I ranked the dog breeds based on the average number of likes the breed recieved, the most liked breeds were the saluki, french bulldog, afghan hound, black and tan coonhound and the flat coated retriever. 

Based on the average rating recieved, the Clumber was the most loved dog breed. 

#### Dog Stages

Personally, the first time I heard the words doggo, puppo, pupper and floofer, was while I worked on this project. They are known as dog stages. Basically they tell you the age group the dog falls into. The same way we have infants, teenagers, adults e.t.c

The WeRateDogs tweet data showed that the most popular dog stage was the **Pupper with 224 tweets relating to it**. 

Based on the number of tweets relating to them, the remaining stages were arranged in this order: doggo, puppo, doggo and pupper, floofer, doggo and floofer, doggo and puppo.





#### Correlation between number of retweets and number of favorites recieved by each tweet

From the chart produced, we can see that there is a **strong positive correlation** between the favorites count and retweet count.
This means that tweets with a high number of retweets also tend to have a high number of favorites. 

This correlation means the WeRateDogs audience likes to show their affection using both retweets and favorites.
